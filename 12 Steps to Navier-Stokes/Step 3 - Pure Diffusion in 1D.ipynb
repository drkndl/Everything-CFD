```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx=41
dx=2/(nx-1)
nt=20
nu=0.3                              #diffusion parameter (viscosity)
sigma=0.2
dt=sigma*dx**2/nu

u=np.ones(nx)
u[int(0.5/dx):int(1/dx+1)]=2
print(u)
```

```python
for it in range(nt):
    un=u.copy()
    for i in range(1, nx-1):
        u[i] = un[i] + nu*dt/(dx)**2*(un[i+1]-2*un[i]+un[i-1])
print(u)
```

```python
plt.plot(np.linspace(0,2,nx),u)
plt.title("Pure Diffusion in 1D")
plt.xlabel("x, m")
plt.ylabel("u")
plt.show()
```
